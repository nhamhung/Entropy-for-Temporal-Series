## Comparative Study:

```python
import entropy as ent
import neurokit2 as nk
import numpy as np
import pandas as pd

df_healthy = pd.DataFrame()
df_unhealthy = pd.DataFrame()
```

```python
from prepare_data import *
eeg_healthy_collections = load_data("../datasets/epilepsy/Z", "Z")
eeg_unhealthy_collections = load_data("../datasets/epilepsy/F", "F")
```

Dataset: http://epileptologie-bonn.de/cms/front_content.php?idcat=193&lang=3 for epilepsy research
- In total, there are 5 sets, each containing 100 single-channel EEG segments.
- Each segment has 4096 samples.
- Data in sets A and B are extracranial EEGs from 5 healthy volunteers with eyes opened and closed respectively.
- Sets C and D are intracranial data over interictal periods
- Set E are over ictal periods.
- Set C, D, E are composed from EEGs of 5 patients.


```python
print("Number of segments per set: ", len(eeg_healthy_collections))
print("Length of each segment: ", len(eeg_healthy_collections['Z1']))
```

```python
plot_segment(eeg_healthy_collections, 'Z', 1, 5)
```

- Plot the first 5 segments of an unhealthy set

```python
plot_segment(eeg_unhealthy_collections, 'F', 1, 5)
```

## Observation:
The healthy segments appear more random. Hence, a higher entropy values would expected.

Compute Entropy metrics for all segments in each set

```python
# Compute the entropy metrics
for key in eeg_healthy_collections.keys(): # append to data frame each sine[i] value with different factors of noise
    df_healthy = df_healthy.append(
                    {
                    'DispersionEnt': ent.dispersion_entropy(eeg_healthy_collections[key], 4, 3, 1),
                    'FuzzyEnt': nk.entropy_fuzzy(eeg_healthy_collections[key], 1, 3),
                    'MultiscaleEnt': nk.entropy_multiscale(eeg_healthy_collections[key], 3),
                    'PermEnt': ent.perm_entropy(eeg_healthy_collections[key], order=3, normalize=True),
                    'SVDEnt' : ent.svd_entropy(eeg_healthy_collections[key], order=3, normalize=True),
                    'SpecEnt' : ent.spectral_entropy(eeg_healthy_collections[key], 100, method='welch', normalize=True),
                    'AppEnt': ent.app_entropy(eeg_healthy_collections[key], order=2),
                    'SampleEnt': ent.sample_entropy(eeg_healthy_collections[key], order=2),
                   }, ignore_index=True)
```

```python
# Compute the entropy metrics
for key in eeg_unhealthy_collections.keys(): # append to data frame each sine[i] value with different factors of noise
    df_unhealthy = df_unhealthy.append({
                    'DispersionEnt': ent.dispersion_entropy(eeg_unhealthy_collections[key], 4, 3, 1),
                    'FuzzyEnt': nk.entropy_fuzzy(eeg_unhealthy_collections[key], 1, 3),
                    'MultiscaleEnt': nk.entropy_multiscale(eeg_unhealthy_collections[key], 3),
                    'PermEnt': ent.perm_entropy(eeg_unhealthy_collections[key], order=3, normalize=True),
                    'SVDEnt' : ent.svd_entropy(eeg_unhealthy_collections[key], order=3, normalize=True),
                    'SpecEnt' : ent.spectral_entropy(eeg_unhealthy_collections[key], 100, method='welch', normalize=True),
                    'AppEnt': ent.app_entropy(eeg_unhealthy_collections[key], order=2),
                    'SampleEnt': ent.sample_entropy(eeg_unhealthy_collections[key], order=2),
                   }, ignore_index=True)
```

# Comparative Study

```python
x = eeg_healthy_collections['Z1']
print("Dispersion Ent: ", ent.dispersion_entropy(x, 4, 3, 1))
print("Fuzzy Ent: ", nk.entropy_fuzzy(x, 1, 3))
print("Multi-scale Ent: ", nk.entropy_multiscale(x, 3))
print("Perm Ent: ", ent.perm_entropy(x, order=3, normalize=True))                 # Permutation entropy
print("Spectral Ent: ", ent.spectral_entropy(x, 100, method='welch', normalize=True)) # Spectral entropy
print("SVD Ent: ", ent.svd_entropy(x, order=3, delay=1, normalize=True))         # Singular value decomposition entropy
print("App Ent: ", ent.app_entropy(x, order=3, metric='chebyshev'))              # Approximate entropy
print("Sample Ent: ", ent.sample_entropy(x, order=3, metric='chebyshev'))           # Sample entropy
```

```python
from timer import Timer
t = Timer()

# Entropy
print("Dispersion Ent:")
t.start()
ent.dispersion_entropy(x, 4, 3, 1)
t.stop()

print("Fuzzy Ent:")
t.start()
nk.entropy_fuzzy(x, 1, 3)
t.stop()

print("Multiscale Ent:")
t.start()
nk.entropy_multiscale(x, 3)
t.stop()

print("Perm Ent:")
t.start()
ent.perm_entropy(x, order=3, delay=1)
t.stop()

print("Spectral Ent:")
t.start()
ent.spectral_entropy(x, 100, method='fft')
t.stop()

print("SVD Ent:")
t.start()
ent.svd_entropy(x, order=3, delay=1)
t.stop()

print("App Ent:")
t.start()
ent.app_entropy(x, order=3) # Slow
t.stop()

print("Samp Ent:")
t.start()
ent.sample_entropy(x, order=3) # Numba
t.stop()

```
